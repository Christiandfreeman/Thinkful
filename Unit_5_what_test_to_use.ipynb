```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = pd.read_csv('ESSdata_Thinkful.csv')
df.head()
```

```python
df.describe()
```

```python
df.count()
```

```python
df.dropna(0,'any',inplace=True)
df.count()
```

```python
df.rename(columns={'idno':'id','agea':'age'}, inplace=True)
df.count()
```

```python
df.cntry.unique()
```

```python
czech = df[df['cntry'] == 'CZ']
denmark = df[df['cntry'] == 'DE']
switzerland = df[df['cntry'] == 'CH']
spain = df[df['cntry'] == 'ES']
norway = df[df['cntry'] == 'NO']
sweden = df[df['cntry'] == 'SE']
spain.head()
```

```python
czech_2012 = czech[czech['year'] == 6]
czech_2014 = czech[czech['year'] == 7]
denmark_2012 = denmark[denmark['year'] == 6]
denmark_2014 = denmark[denmark['year'] == 7]
switzerland_2012 = switzerland[switzerland['year'] == 6 ]
switzerland_2014 = switzerland[switzerland['year'] == 7]
spain_2012 = spain[spain['year'] == 6]
spain_2014 = spain[spain['year'] == 7]
norway_2012 = norway[norway['year'] == 6 ]
norway_2014 = norway[norway['year'] == 7]
sweden_2012 = sweden[sweden['year']== 6]
sweden_2014 = sweden[sweden['year'] == 7]
sweden_2014.head()
```

```python
czech_2012_sum = czech_2012['ppltrst'].sum()
czech_2014_sum = czech_2014['ppltrst'].sum()
denmark_2012_sum = denmark_2012['ppltrst'].sum()
denmark_2014_sum = denmark_2014['ppltrst'].sum()
switzerland_2012_sum = switzerland_2012['ppltrst'].sum()
switzerland_2014_sum = switzerland_2014['ppltrst'].sum()
spain_2012_sum = spain_2012['ppltrst'].sum()
spain_2014_sum = spain_2014['ppltrst'].sum()
norway_2012_sum = norway_2012['ppltrst'].sum()
norway_2014_sum = norway_2014['ppltrst'].sum()
sweden_2012_sum = sweden_2012['ppltrst'].sum()
sweden_2014_sum = sweden_2014['ppltrst'].sum()

```

```python
print('Total loss or gain from 2012-2014 by country: ','\nCH:',
      czech_2012_sum - czech_2014_sum,
     '\nDE: ',denmark_2012_sum - denmark_2014_sum,
     '\nSW: ',switzerland_2012_sum - switzerland_2014_sum,
     '\nSP: ',spain_2012_sum - spain_2014_sum,
     '\nNO: ',norway_2012_sum - norway_2014_sum,
     '\nSE: ',sweden_2012_sum - sweden_2014_sum)






```

```python
sns.lineplot('cntry', 'ppltrst', hue='year', data=df)
plt.legend(loc='lower right')
plt.show()
```

```python
czech_happy_sum = czech_2012['happy'].sum() - czech_2014['happy'].sum()
denmark_happy_sum = denmark_2012['happy'].sum() - denmark_2014['happy'].sum()
switzerland_happy_sum =switzerland_2012['happy'].sum() - switzerland_2014['happy'].sum()
spain_happy_sum = spain_2012['happy'].sum() - spain_2014['happy'].sum()
norway_happy_sum = norway_2012['happy'].sum() - norway_2014['happy'].sum()
sweden_happy_sum = sweden_2012['happy'].sum() - sweden_2014['happy'].sum()
```

Looks like peoples trust stays fairly consistant over the years. There was a slight increase in some countries and a slight decrease in others. 

```python
print('Happiness by country: ', '\nCZ: ', czech_happy_sum,
     '\nDE: ', denmark_happy_sum,
     '\nCH: ', switzerland_happy_sum,
     '\nES: ', spain_happy_sum,
     '\nNO: ', norway_happy_sum,
     '\nSE: ', sweden_happy_sum)
```

```python
sns.barplot('gndr', 'age', hue= 'tvtot', data = df)
plt.show()
```

```python
males = df[df['gndr'] == 1.0]
females = df[df['gndr'] == 2.0 ]
male_tv_total = males['tvtot'].sum()
female_tv_total = females['tvtot'].sum()

print('Male TV Watch Time: ', male_tv_total,
     '\nFemale TV Watch Time: ', female_tv_total)
```

Looks like Women Watch more tv than men by a small margin 

```python
df_partner = df[df['year'] == 6] 
with_partner = df_partner[(df_partner['partner'] == 1 )]
without_partner = df_partner[df_partner['partner'] == 2]
with_part_fair = with_partner['pplfair'].sum()
without_part_fair = without_partner['pplfair'].sum()
print('Fair Sum for people who live with their partner: ', with_part_fair,
     '\nFair sum for people who do not live with their partner: ', without_part_fair)
```

Looks like people who live with their partner think people are more fair than people who do not live with their partner
